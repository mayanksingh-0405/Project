Importing all the required libraries

```python
import numpy as np # Calculations
import pandas as pd # Data Preprocessing, CSV file (I/O)
import matplotlib.pyplot as plt 
import seaborn as sns
%matplotlib inline 
```

```python
# importing dataset into pandas 
goldstock_df = pd.read_csv('goldstock.csv')
```

```python
goldstock_df
```

For this, we can see that there is an unnecessary unamed column. We will have to remove it. 

```python
goldstock_df = goldstock_df.iloc[:, 1:]
```

```python
goldstock_df
```

Now we have removed the useless column from the dataframe. Let's try to get some basic information about the dataframe.

```python
goldstock_df.info()
```

We should convert the Data object to datetime for easier.

```python
goldstock_df['Date'] = pd.to_datetime(goldstock_df['Date'])
goldstock_df.info()
```

```python
goldstock_df.describe()
```

```python
# Find the number of null values in the dataset
goldstock_df.isna().sum()
```
